# Single Variant Analysis

```python
df_Mobile['Mobile Colour'].value_counts().head(10)
```

```python
df_Mobile['Mobile Colour'].value_counts().head(10).plot.pie(autopct="%.1f%%")
plt.title('Mobile Colour')
plt.show()
```

# This pie chart shows the details of top 10 high demanding colour of the phones in the market where I found that 'Dark Purple' is in high demand

```python
df_Mobile['Mobile Brand'].value_counts()
```

```python
df_Mobile['Mobile Brand'].value_counts().plot.barh()
plt.title('Mobile Brand')
plt.show()
```

## This graph shows that 'SAMSUNG' is the most demanding mobile brand in the market and after that four more brands are in demand 'REALME', 'APPLE', 'VIVO', 'REDME'.

```python
df_Mobile.boxplot(column='Rating')
plt.title('Rating')
plt.show()
```

```python
df_Mobile['Rating'].value_counts()
```

## Most of the phones got 4.2, 4.3 and 4.4 so it means quality of the product is good and customers are well satisfied.

```python
df_Mobile.boxplot(column='Discount')
plt.title('Discount')
plt.show()
```

## In this Discount column there are few outliers but most of the brand gives discount between 10% to 25% 

```python
sns.boxplot(x='Actual Price',data=df_Mobile)
plt.title('Actual Price')
plt.show()
```

# Most of the porduct is between ₹15000 to ₹50000 Because demand of the market is high

```python
sns.heatmap(df_Mobile[['Mobile RAM','Mobile ROM','Mobile Battery','Rating','Actual Price','Discount','Selling Price']].corr(),annot=True)
plt.show()
```

# 1. Mobile ROM is having a medium correlation with Mobile RAM.
# 2. Mobile Battery is having a low negative correlation with Mobile RAM and Mobile ROM.
# 3. Rating is having a low negative correlation with Mobile RAM, low positive correlation with Mobile ROM and Mobile Battery.
# 4. Actual Price is having medium Positive correlation with Mobile RAM and Mobile ROM, low positive correlation with Rating and medium negative correlation with Mobile Battery.
# 5. Discount is having a low negative correlation with Mobile ROM, MObile RAM, Mobile Battery, Rating, Actual Price.
# 6. Selling price is having a medium positive correlation with Mobile RAM, Mobile ROM and Rating, medium negative correlation with Mobile Battery and discount, best correlation with Actual price

```python
sns.pairplot(data=df_Mobile,vars=['Mobile RAM','Mobile ROM','Mobile Battery','Rating','Actual Price','Discount','Selling Price'])
plt.show()
```
