```python
MobileName=[]
Feature=[]
Rating=[]
ActualPrice=[]
Discount=[]
SellingPrice=[]
Page_no=[]

for i in range(1,51):
    start_time=time.time()
    url3='https://www.flipkart.com/search?q=mobile+5g+under+25000&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={}'.format(i)
    page=requests.get(url3)
    pagecontent=page.text
    soup=BeautifulSoup(pagecontent)
    
    for x in soup.findAll('div',attrs={'class':'_3pLy-c row'}):
        mob_name = x.find('div',attrs={'class':'_4rR01T'})
        mob_feat = x.find('ul',attrs={'class':'_1xgFaf'})
        mob_rate = x.find('div',attrs={'class':'_3LWZlK'})
        mob_acttualprice = x.find('div',attrs={'class':'_3I9_wc _27UcVY'})
        dis = x.find('div',attrs={'class':'_3Ay6Sb'})
        mob_sellprice = x.find('div',attrs={'class':'_30jeq3 _1_WHN1'})
    
        if mob_name is None:
            MobileName.append(np.NaN)
        else:
            MobileName.append(mob_name.text)
    
        if mob_feat is None:
            Feature.append(np.NaN)
        else:
            Feature.append(mob_feat.text)
    
        if mob_rate is None:
            Rating.append(np.NaN)
        else:
            Rating.append(mob_rate.text)
    
        if mob_acttualprice is None:
            ActualPrice.append(np.NaN)
        else:
            ActualPrice.append(mob_acttualprice.text)
    
        if dis is None:
            Discount.append(np.NaN)
        else:
            Discount.append(dis.text)
    
        if mob_sellprice is None:
            SellingPrice.append(np.NaN)
        else:
            SellingPrice.append(mob_sellprice.text)
    
        Page_no.append(i)
    print('Page {} completed in {}'.format(i,time.time()-start_time))
```

```python
print(len(MobileName))
print(len(Feature))
print(len(Rating))
print(len(ActualPrice))
print(len(Discount))
print(len(SellingPrice))
print(len(Page_no))
```
